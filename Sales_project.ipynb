```python
# Identify average customer visit in the type B store in April Months
#sale table
sale_type_b_store=sales[sales["Store"].isin(store_df[store_df["Type"]=="B"]['Store'])]# retrive store which has only type B
sales_april=sale_type_b_store[sale_type_b_store["Date"].dt.month==4]# retrive april month
total_sale_typebstore=sales_april.groupby("Store")["Weekly_Sales"].sum().to_frame() #grouping store and calculating the total sales for april month
# feature table
features1_type_b_store=features_1[features_1["Store"].isin(store_df[store_df["Type"]=="B"]['Store'])]# retrive store which has only type b in feature table
features1_april=features1_type_b_store[features1_type_b_store["Date"].dt.month==4]# retrive april month
total_feature1_typebstore=features1_april.groupby("Store")["CPI"].sum().to_frame()#grouping store and calculating the total sales for april month
#merge
merge_table=pd.merge(total_sale_typebstore,total_feature1_typebstore, on="Store")
merge_table["Average_customer_visit"]=merge_table["Weekly_Sales"]/merge_table["CPI"]
merge_table
```

```python
# â— Identify best average sales in holiday week for all store types
Sales_Isholiday=sales[sales["IsHoliday"]==True]
sales_merge_table=pd.merge(Sales_Isholiday,store_df, on="Store")
Average_sales=sales_merge_table.groupby("Type")["Weekly_Sales"].mean().to_frame()
Average_sales.sort_values(by=["Weekly_Sales"],ascending=False).head(1)
```

```python
# â— Which store had a worst sales in leap year
sale_leapyear=sales[sales["Date"].dt.year==2012]
sum_sales_leapyear=sale_leapyear.groupby('Store')["Weekly_Sales"].sum().to_frame()
sum_sales_leapyear.sort_values(by=["Weekly_Sales"],ascending=True).head(1)
```

```python
# â— What is the expected sales of each department when unemployment factor is greater>8
```

```python
features_1[features_1["Unemployment"]>8].isnull().sum()
```

```python
unemployment_factor=features_1[features_1["Unemployment"]>8]
unemployment_factor["totalMark"]=unemployment_factor[["MarkDown1","MarkDown2","MarkDown3","MarkDown4","MarkDown5"]].sum(axis=1)
unemployement_total_mark=unemployment_factor[["Store","Date","totalMark"]]
merge_table1=pd.merge(sales,unemployement_total_mark,on=("Store","Date"))
merge_table1["expected_sales"]=merge_table1["Weekly_Sales"]+merge_table1["totalMark"]
merge_table1.groupby("Dept")["expected_sales"].sum().to_frame()
```

```python
 #Aggregate the net(total) sales of each department on month wise
sales["month"]=sales["Date"].dt.month
sales.groupby(['Dept','month'])["Weekly_Sales"].sum().to_frame()
```

```python
# Which store performs high sales in week wise
sales["Week"]=sales["Date"].dt.strftime('%U')
high_week=sales.groupby(['Store','Week'])["Weekly_Sales"].sum().to_frame().reset_index()
high_week.sort_values(by=["Weekly_Sales"],ascending=False).head(10)
```

```python
#Identify better department performance based on the store on all the week
sal=sales.groupby(['Store','Dept'])["Weekly_Sales"].sum().to_frame()
sal.sort_values(by="Weekly_Sales",ascending=False).head(10)
```

```python
# â— Identify the store which has minimum fuel price based on the week >>>>
features_1["Week"]=features_1["Date"].dt.strftime('%U')
```

```python
sal1=features_1.groupby(['Store','Week'])["Fuel_Price"].min().to_frame()
sal1.sort_values(by="Fuel_Price",ascending=True).head(10)
```

```python
 #Identify overall performance of the store based on year wise
sales["year"]=sales["Date"].dt.year
high_year=sales.groupby(['Store','year'])["Weekly_Sales"].sum().to_frame().reset_index()
high_year.sort_values(by=["Weekly_Sales"],ascending=False).head(10)
```
